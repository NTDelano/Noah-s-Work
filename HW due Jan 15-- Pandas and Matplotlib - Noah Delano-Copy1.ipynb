## Question 2

Create a DataFrame that records the number of hours you slept each day last week and the number of hours you studied each day last week. Index your DataFrame by ['Monday','Tuesday, etc]

```python
Days_Of_The_Week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
Hours_Spent = ['Hours Studying', "Hours Slept"]

Monday = pd.Series([6, 6], index = Hours_Spent)
Tuesday = pd.Series([12, 7.5], index = Hours_Spent)
Wednesday = pd.Series([15, 4.6], index = Hours_Spent)
Thursday = pd.Series([7.4, 7.6], index = Hours_Spent)
Friday = pd.Series([6.6, 8.3], index = Hours_Spent)
Saturday = pd.Series([3, 10.1], index = Hours_Spent)
Sunday = pd.Series([8, 9.8], index = Hours_Spent)

Overall_Hours_Spent = pd.DataFrame([Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday], index = Days_Of_The_Week)
Overall_Hours_Spent
```

## Question 3

Create a scatter plot for number of hours of sleep (x-axis) versus the number of hours of studying (y-axis). 

```python
import matplotlib.pyplot as plt 

x=Overall_Hours_Spent['Hours Slept']
y=Overall_Hours_Spent['Hours Studying']
plt.scatter(x,y)
plt.title("Scatterplot of Hours Spent Sleeping VS. Studying")
plt.xlabel("Hours Slept")
plt.ylabel("Hours Studying")
plt.show()
```

## Question 4

Create a bar plot for number of hours of studying last week 

```python
Days_Of_The_Week_Condensed = ['M', 'T', 'W', 'Th', 'F', 'S', 'Su']
Studying = Overall_Hours_Spent['Hours Studying'].values
plt.bar(Days_Of_The_Week_Condensed, Studying,width=.75,color='blue')
plt.title("Hours Spent Studying Each Day Last Week")
plt.show()
```

The following code loads the olympics dataset (olympics.csv), which was derrived from the Wikipedia entry on [All Time Olympic Games Medals](https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table), and does some basic data cleaning. 

The columns are organized as # of Summer games, Summer medals, # of Winter games, Winter medals, total # number of games, total # of medals. Use this dataset to answer the questions below.

```python
url='http://facweb1.redlands.edu/fac/Tamara_Veenstra/ML/olympics.csv'
df = pd.read_csv(url, index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)
        
df.drop('Totals', inplace=True)

df.head(160)
```
