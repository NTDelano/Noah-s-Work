Use sklearn to redo your linear regression from HW due Jan 27, Question 3. That is:

```python
# Here's data that you used before.
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
URL = 'http://facweb1.redlands.edu/fac/joanna_bieri/machinelearning/LifeExpectancyData.csv'
df = pd.read_csv(URL)
#I Know I could have done this simpler, but YOLO am I right?
df.drop('Country', axis=1, inplace=True)
df.drop('Year', axis=1, inplace=True)
df.drop('Status', axis=1, inplace=True)
df.drop('Infant deaths', axis=1, inplace=True)
df.drop('Alcohol', axis=1, inplace=True)
df.drop('percentage expenditure', axis=1, inplace=True)
df.drop('Hepatitis B', axis=1, inplace=True)
df.drop('Measles ', axis=1, inplace=True)
df.drop('under-five deaths ', axis=1, inplace=True)
df.drop('Polio', axis=1, inplace=True)
df.drop('Total expenditure', axis=1, inplace=True)
df.drop('Diphtheria ', axis=1, inplace=True)
df.drop('HIV/AIDS', axis=1, inplace=True)
df.drop('GDP', axis=1, inplace=True)
df.drop('Population', axis=1, inplace=True)
df.drop('thinness  1-19 years', axis=1, inplace=True)
df.drop('thinness 5-9 years', axis=1, inplace=True)
df.drop('BMI', axis=1, inplace=True)

df.dropna(inplace=True)

df.describe()
```

```python
y = df['Life expectancy '] #Sets dependent values to life expectancy

AMortality = df.iloc[:,1].values.reshape(-1,1) #Column for Adult Mortality

Income = df.iloc[:,2].values.reshape(-1,1) #Column for Income

Schooling = df.iloc[:,3].values.reshape(-1,1) #Column for Schooling

x = np.hstack((AMortality, Income, Schooling)) #Places the three previous columns into one matrix
print(x.shape, y.shape)
sns.pairplot(df,height=2.5)
```

```python
SScalar=StandardScaler() #Applied standard scaler from #1
SScalar.fit(x)
xSS2 = SScalar.transform(x)
print(xSS2.shape)
```

```python
lin_reg = LinearRegression() #Establish linear regression function

lin_reg.fit(xSS2,y) #Appliy linear regression to our normalized x and y

#Theta will be = [lin_reg.intercept_, lin_reg.coef_]
thetas = [lin_reg.intercept_, lin_reg.coef_[0], lin_reg.coef_[1], lin_reg.coef_[2]] 
print(thetas)
```

```python
lin_reg.predict(xSS2)
from sklearn.metrics import mean_squared_error
```
